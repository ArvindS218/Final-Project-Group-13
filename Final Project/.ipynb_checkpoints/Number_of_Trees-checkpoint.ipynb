Submetric: number of trees in each neighbourhood.
The more the trees, the more easier it is to start a fire and the more widespread the fire is.
The database link: https://data.wprdc.org/dataset/city-trees/resource/1515a93c-73e3-4425-9b35-1cd11b2196da

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv('trees.csv')
print(df['neighborhood'].value_counts().head(45709))
print()
print()
df['neighborhood'].value_counts().plot(kind='bar',figsize=(15, 10))
plt.xlabel("Neighborhoods")
plt.ylabel("Number of trees")
plt.tick_params(axis='x', which='major', pad=10)




```

Conclusion: Only considering the number of trees submetric, Squirrel Hill South would be the best and easiest neighborhood to start a widespreading fire.
