```python
#query_mask = traffic_data['average_daily_car_traffic']!=nan
#graph = pd.DataFrame(columns=traffic_data['average_daily_car_traffic'],index=traffic_data['neighborhood'])
#graph.plot(kind='bar')

query_mask = traffic_data['average_daily_car_traffic']>7000
hehehawhaw = traffic_data[query_mask]
#hehehawhaw.head(40)
final = dict()
for row in hehehawhaw[1:]:
    if row[18] not in final:
        final[row[18]] = row[7]
    else:
        final[row[18]] = final[row[18]]+row[7]
final 

        
        
#hehehawhaw['average_daily_car_traffic'].plot(x='neighborhoods')
```
